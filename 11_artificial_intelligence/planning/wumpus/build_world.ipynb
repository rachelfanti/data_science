```python
import itertools

def world_builder(world):
    build_dict = {
        'parent': 'define',
        'blocks': ['problem', ':domain', ':objects', ':init', ':goal', ':metric'],
        'domain': 'wumpus',
        ':objects': ['gold', 'arrow', 'agent', 'wumpus']
    }
    
    res = '(' + build_dict['parent'] + '\n'
    tile_list = [(col, row) for col in range(world['size'][0]) for row in range(world['size'][1])]

    for block in build_dict['blocks']:
        res += '\t(' + block + ' '

        if block == 'problem': 
            res += world['id']
        elif block == ':domain':
            res += build_dict['domain']
        elif block == ':objects':
            res += '\n'
            for tile in tile_list:res += '\t\tsq-' + str(tile[0]) + '-' + str(tile[1]) + '\n'
            for object in build_dict[':objects']: res += '\t\t' + object + '\n'
        elif block == ':init':
            res += '\n'
            for a, b in itertools.combinations(tile_list, 2):
                curr_diff = tuple(map(lambda i, j: abs(i - j), a, b))
                if  curr_diff == (0,1) or curr_diff == (1,0): 
                    res += '\t\t(adj sq-' + str(a[0]) + '-' + str(a[1]) + ' sq-' + str(b[0]) + '-' + str(b[1]) + ')\n'
                    res += '\t\t(adj sq-' + str(b[0]) + '-' + str(b[1]) + ' sq-' + str(a[0]) + '-' + str(a[1]) + ')\n'
            
            if world['pits']:
                res += '\n'
                for pit in world['pits']:
                    res += '\t\t(pit sq-' + str(pit[0]) + '-' + str(pit[1]) + ')\n'

            res += '\n'
            for object, loc in zip(['gold', 'agent', 'wumpus'], [world['golds'][0], (0, 0), world['wumpuses'][0]]):
                res += '\t\t(is-' + object + ' ' + object + ')\n\t\t(at ' + object + ' sq-' + str(loc[0]) + '-' + str(loc[1]) + ')\n\n'

            res += '\t\t(wumpus-in sq-' + str(world['wumpuses'][0][0]) + '-' + str(world['wumpuses'][0][1]) + ')\n\n'
            res += '\t\t(is-arrow arrow)\n\t\t(have agent arrow)\n\n'
            res += '\t\t(= (total-cost) 0)\n'
        elif block == ':goal':
            res += '(and\n\t\t\t(have agent gold)\n\t\t\t(at agent sq-' + str(world['exits'][0][0]) + '-' + str(world['exits'][0][1]) + ')\n\t\t)\n'
        elif block == ':metric':
            res += '\n\t\tminimize (total-cost)\n'
        res += '\t)\n'
    res += ')'

    text_file = open(world['id'] + '.pddl', "w")
    text_file.write(res)
    text_file.close()   
```

```python
world_01 = {
  "id": "w01",
  "size": (10, 10),
  "hunters": [(0, 0, "N")],
  "pits": [],
  "wumpuses": [(0, 9)],
  "exits": [(0, 0)],
  "golds": [(5, 5)]
}
world_builder(world_01)
!fast-downward wumpus.pddl w01.pddl --heuristic "h=blind()" --search "eager_greedy([h])"
!fast-downward wumpus.pddl w01.pddl --heuristic "h=ff()" --search "eager_greedy([h])"
!fast-downward wumpus.pddl w01.pddl --heuristic "h=add()" --search "eager_greedy([h])"
```
